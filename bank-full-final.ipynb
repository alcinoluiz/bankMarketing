# Features & Estimators

```python
X = df_balanced[['age', 'default', 'balance', 'housing',
       'loan', 'contact', 'day', 'duration', 'campaign', 'pdays',
       'previous', 'poutcome']]
```

```python
y = df_balanced['y_bin']
```

```python
X.loc[:,'poutcome_success'] = X.poutcome.apply(lambda x: 1 if x == 'success' else 0)
X.loc[:,'credit_default'] = X.default.apply(lambda x: 1 if x == 'yes' else 0)
X.loc[:,'housing_loan'] = X.housing.apply(lambda x: 1 if x == 'yes' else 0)
X.loc[:,'personal_loan'] = X.loan.apply(lambda x: 1 if x == 'yes' else 0)
X.loc[:,'contact_cellular'] = X.contact.apply(lambda x: 1 if x == 'cellular' else 0)
X.loc[:,'first_contact'] = X.pdays.apply(lambda x: 1 if x == -1 else 0)
```

```python
X
```

```python
features_cat = ['poutcome_success', 'credit_default', 'housing_loan', 'personal_loan', 'contact_cellular', 'first_contact']
```

```python
features_num = ['age', 'balance', 'duration', 'campaign', 'previous']
```

```python
features_all = features_cat + features_num
features_all
```

```python
X = X[features_all]
```

```python
X.shape
```

```python
X.corr()['credit_default']
```

```python
X.corr()['housing_loan']
```

```python
y.shape
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)

for train_index, test_index in skf.split(X[features_all], y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
print(len(X_train))
print(len(X_test))
```

```python
estimators_n = ['RandomForestClassifier', 'LinearSVC', 'GradientBoostingClassifier']
```

```python
estimators = []
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import cross_val_score
```

```python
estimators.append(RandomForestClassifier(max_depth=2, random_state=42))
```

```python
estimators.append(GradientBoostingClassifier(max_depth=2, random_state=42))
```

```python
estimators.append(LinearSVC(random_state=42, tol=1e-5))
```

```python
for i, estimator in enumerate(estimators):
    print(estimators_n[i])
    estimator.fit(X_train.values, y_train.values);
    print('score:', estimator.score(X_train, y_train))
    cv_score = cross_val_score(estimator, X_test, y_test, cv=10)
    print(cv_score)
    print('cv score mean:', cv_score.mean())
    print()
```

### Fazendo uma relação entre número de contatos e sucesso da campanha quais são os pontos relevantes a serem observados?
- Maioria das campanhas que ocorreram tiveram mais falhas que sucesso
- O sucesso tende a diminuir conforme o número de contatos
- Maior parte dos contatos foram feitos via celular
- Existem muitos resultados 'deconhecidos'
- Maio foi o mês com mais contatos
- Apenas 11% dos contatos assinaram o termo de depósito
- 84% das adesões aconteceram nas primeiras 3 ligações, sendo 46% delas só na primeira 

### Baseando-se nos resultados de adesão desta campanha qual o número médio e o máximo de ligações que você indica para otimizar a adesão?
- 84% das adesões aconteceram nas primeiras 3 ligações, sendo 46% delas só na primeira 
- A tendência é que na primeira ligação ocorra a maioria das adesões
- Podemos deixar a média de 2 ligações
- Podemos deixar no máximo 6 ligações

### O resultado da campanha anterior tem relevância na campanha atual?

- 64% dos contatos de sucesso anterior conseguem adesão
Então podemos dizer que sim, mesmo que muitos valroes estão marcados como 'desconhecido'.
